```python
def compute_distance(phm_drop):

  from re import I
  sum = 0

  o = 0
  d = 0

  while True:
      #No patient at that o,d combo
      if phm_drop.iloc[o,d] == 0:

        #If we are at the end of the columns
        if o >= (phm_drop.shape[1] - 1):
          #If we are also at the end of rows, we are done
          if d >= (phm_drop.shape[0] - 1):
            break

          else:
          #At end of columns but not at the end of the rows, so move down a row
            o = 0
            d = d + 1
            continue

        else:
        #If not at the end of the columns, simply move a column
          o = o + 1
          continue

      else:
      #Patient at that o,d combo


        phm_drop.iloc[o,d] = phm_drop.iloc[o,d] - 1
        sum = sum + avg_distance

        #Iterate through the row corresponding to that previous destination's column
        #See if there is a patient waiting to be picked up from the destination
        i = d
        j = 0

        while True:
          #No patient at that i,j combo
          if phm_drop.iloc[i, j] == 0:
            #If we are at the end of the columns
            if j >= (phm_drop.shape[1] - 1):
              sum = sum + avg_distance  #There was no patient waiting at that destination, so we have to travel to the next hospital (penalty)
              break

            #Not at the end of the columns, simply move a column
            else:
              j = j + 1
              continue

          else:
            #Yes patient at that i,j combo
            phm_drop.iloc[i,j] = phm_drop.iloc[i,j] - 1 #Remove the patients from the demand list
            sum = sum + avg_distance
            break

        #Overall while loop
        continue

  return (sum)
```

```python
phm_sub_1_dist = compute_distance(phm_mod_subs1)

phm_sub_2_dist = compute_distance(phm_mod_subs2)

phm_sub_3_dist = compute_distance(phm_mod_subs3)

phm_sub_4_dist = compute_distance(phm_mod_subs4)
```
